```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/flipkart_com-ecommerce_sample.csv')
```

```python
# Convert 'Order Date' to datetime
data['crawl_timestamp'] = pd.to_datetime(data['crawl_timestamp'])
daily_sales = data.groupby(data['crawl_timestamp'].dt.date)['discounted_price'].sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(daily_sales.index, daily_sales.values, marker='o')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Calculate total sales
total_sales = data['discounted_price'].sum()
```

```python
# Print the result
print("Total Sales:", total_sales)
```

```python
# Create a bar chart
plt.bar(['Total Sales'], [total_sales])

# Add labels and title
plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Total Sales")

# Display the chart
plt.tight_layout()
plt.show()
```

```python
# Group by product name and sum quantities
product_sales = data.groupby('product_name')['discounted_price'].sum()

# Sort by sales in descending order
best_selling_products = product_sales.sort_values(ascending=False)

# Display the top 10 best-selling products
print(best_selling_products.head(10))
```

```python
# Assuming you have already calculated 'best_selling_products' as in the previous response

# Get the top 10 product names and their sales
top_products = best_selling_products.head(10)
product_names = top_products.index
sales_values = top_products.values

# Create a bar chart
plt.figure(figsize=(12, 12))  # Adjust figure size for better readability
plt.bar(product_names, sales_values)

# Add labels and title
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.title("Top 10 Best-Selling Products")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Display the chart
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Calculate total spending per customer
customer_spending = data.groupby('pid')['discounted_price'].sum().sort_values(ascending=False).head(10)
```

```python
# Plot top 10 customers by spending
plt.figure(figsize=(10, 6))
customer_spending.plot(kind='bar', color='coral')
plt.title('Customers by Spending')
plt.xlabel('CustomerID')
plt.ylabel('Total Spending')
plt.xticks(rotation=45)
plt.show()
```
